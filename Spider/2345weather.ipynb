```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.select import Select
import time
import sys
import pandas as pd
from io import StringIO
```

```python
ua = 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.119 Safari/537.36'
HEADERS = {'user-agent':ua}
```

```python
url = 'http://tianqi.2345.com/wea_history/58321.htm'
```

```python
options = webdriver.ChromeOptions() # 创建driver的参数，可选。
options.add_argument('--headless')
options.add_argument('--disable-gpu')
options.add_argument('--ignore-certificate-errors')
driver = webdriver.Chrome(chrome_options=options)
driver.get(url)
#driver.maximize_window()
#driver.find_element_by_css_selector('#switchHisCity').click()
time.sleep(1)
df = pd.DataFrame(columns=['date','a','b','c','d','e','f'])
province = Select(driver.find_element_by_css_selector("#selectProv"))
count_province = len(province.options)
print('省份个数：{0}'.format(count_province))
print('----------------start---------------')
print('省ID 城市ID 区ID')
driver.quit()
for i in range(count_province):
    options = webdriver.ChromeOptions() # 创建driver的参数，可选。
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument('--ignore-certificate-errors')
    driver = webdriver.Chrome(chrome_options=options)
    driver.get(url)
    
    driver.find_element_by_css_selector('#switchHisCity').click()
    province = Select(driver.find_element_by_css_selector("#selectProv"))
    province.select_by_index(i)
    city = Select(driver.find_element_by_css_selector("#chengs_ls"))
    count_city = len(city.options)
    print('城市个数：{0}'.format(count_city))
    driver.find_element_by_css_selector('#buttonsdm_dz').click()
    for j in range(count_city):
        driver.find_element_by_css_selector('#switchHisCity').click()
        province = Select(driver.find_element_by_css_selector("#selectProv"))
        province.select_by_index(i)
        city = Select(driver.find_element_by_css_selector("#chengs_ls"))
        city.select_by_index(j)
        district = Select(driver.find_element_by_css_selector("#cityqx_ls"))
        count_district = len(district.options)
        print('区个数：{0}'.format(count_district))
        driver.find_element_by_css_selector('#buttonsdm_dz').click()
        for d in range(count_district):
            options_temp = webdriver.ChromeOptions() # 创建driver的参数，可选。
#             options_temp.add_argument('--headless')
#             options_temp.add_argument('--disable-gpu')
#             options_temp.add_argument('--ignore-certificate-errors')
            driver_temp = webdriver.Chrome(chrome_options=options_temp)
            driver_temp.get(url)
            print(i,j,d)
            
            try:
                driver_temp.find_element_by_css_selector('#switchHisCity').click()
                province_temp = Select(driver_temp.find_element_by_css_selector("#selectProv"))
                province_temp.select_by_index(i)
                city_temp = Select(driver_temp.find_element_by_css_selector("#chengs_ls"))
                city_temp.select_by_index(j)
                district_temp = Select(driver_temp.find_element_by_css_selector("#cityqx_ls"))
                district_temp.select_by_index(d)
            except:
                print('异常：{0},{1},{2}'.format(i,j,d))
            driver_temp.find_element_by_css_selector('#buttonsdm_dz').click()#点击确定
            driver_temp.find_element_by_css_selector("#chooseHisYear > div").click()#选择年
            driver_temp.find_element_by_css_selector('#chooseHisYear > ul > li:nth-child(5)').click()#选年
            time.sleep(1)

            for p in range(2,14):
                driver_temp.find_element_by_css_selector("#chooseHisMonth > div").click()
                driver_temp.find_element_by_css_selector('#chooseHisMonth > ul > li:nth-child({0})'.format(p)).click()
                #time.sleep(1)
                tb = driver_temp.find_element_by_css_selector("#weather_tab > table > tbody")
                c = driver_temp.find_element_by_css_selector('#wrap > div.history-page.clearfix > div.col-1 > div > div.btitle.clearfix > h1')
                s1 = StringIO(tb.text)
                temp = pd.read_csv(s1,sep=' ',encoding='utf8',header=None)
                temp.columns = ['date','a','b','c','d','e']
                temp['f'] = c.text
                df = df.append(temp,ignore_index=True)
            driver_temp.quit()
    driver.quit()
#time.sleep(3)
#driver.quit()
print('---------end-----------')
```
